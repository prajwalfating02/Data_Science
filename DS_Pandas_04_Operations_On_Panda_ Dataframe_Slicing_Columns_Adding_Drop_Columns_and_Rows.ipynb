# Indexing & Selection

```python
import pandas as pd
import numpy as np
```

```python
np.random.randn(10,5)
```

```python
my_dataframe = pd.DataFrame(np.random.randn(10,5), index = 'row1 row2 row3 row4 row5 row6 row7 row8 row9 row10'.split(), columns = 'column1 column2 column3 column4 column5'.split())
```

```python
my_dataframe
```

```python
my_dataframe['column2']                    # Selecting a single column
```

```python
my_dataframe[['column2','column3','column4']]           # Selecting multiple columns     NB : Use either single ' ' OR double "" either works fine 
```

## **Let's add a new column to our DataFrame**

```python
my_dataframe['column6']=my_dataframe ['column1']*2
```

```python
my_dataframe
```

```python
type(my_dataframe['column6'])     # NB : each column is a series, so the dataframe is a table consisting of several series put together
```

## **Dropping Columns and Rows**

```python
my_dataframe.drop('column4',axis=1)    # Use axil=1 to refer to the column, or axis=0 for the row
```

```python
my_dataframe # our dataframe still has the column4
```

```python
my_dataframe.drop('column4',axis = 1, inplace=True)         #Use inplace=True to make the changes affect our original dataset
```

```python
my_dataframe
```

```python
my_dataframe.drop('row2',axis=0,inplace=True)      # Use axis=0 to refer to the rows
```

```python
my_dataframe
```

```python
my_dataframe.loc['row7','column2']                        # print 1st row and 2nd column element
```
