**Difference Between isnull() and isna()**

The two DataFrame methods do exactly the same thing! Even their docs are identical. You can even confirm this in pandas code.

But why have two methods with different names do the same thing? This is beacuse pandas DataFrame are based on R's DataFrame. In R **na** and **null** are two separate things.

However, in python, pandas is built in top of numpy, which has neither **na** nor **null** values. Instead numpy has NaN values (Which stand for "NOT a NUMBER"). Consequently, pandas also uses NaN values.

In short To detect NaN values numpy used **np.isnan().**

To detect NaN values pandas uses either **.isna()** or **.isnull()**. The NaN values are inherited from the fact thet pandas is built on top of numpy, while the two functions names originate from R's DataFrame, whose structure and functionality pandas tried to mimic.
