# Reset Index

```python
import pandas as pd
import numpy as np
```

```python
np.random.randn(10, 5)           # 10*5
```

```python
my_dataframe = pd.DataFrame(np.random.randn(10,5),index='row1 row2 row3 row4 row5 row6 row7 row8 row9 row10'.split(), columns='column1 column2 column3 column4 column5'.split())
```

```python
my_dataframe
```

```python
my_dataframe.index
```

```python
my_dataframe['spin'] = ['sp1', 'sp2', 'sp3', 'sp4', 'sp5', 'sp6', 'sp7', 'sp8', 'sp9', 'sp10']
```

```python
my_dataframe
```

```python
my_dataframe.set_index('spin',inplace=True)     # Use set_index() to set new index
```

```python
my_dataframe
```

```python
my_dataframe.rename(columns={'column1':'first',
                           'column2':'second',
                           'column3':'third',
                           'column4':'fourth',
                           'column5':'fifth'},
                   inplace=True)
```

```python
my_dataframe
```

```python
my_dataframe.rename(columns={'first':1})
```
